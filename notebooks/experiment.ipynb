```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO
```

```python
filepath = "results/resnet18_resize-256_crop-256/master_results.csv"

# Read the data into a pandas DataFrame
df = pd.read_csv(filepath)

# # Sort by F1-score for clearer visualization
# df_sorted = df.sort_values('f1_score', ascending=False)

print("Data loaded successfully:")
# print(df_sorted.head())

# Set the plot style and size
sns.set_style("whitegrid")
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot F1-Score as bars
color = 'tab:blue'
ax1.set_title('Overall Performance by Class', fontsize=16, pad=20)
ax1.set_xlabel('Class Name')
ax1.set_ylabel('F1 Score (Bars)', color=color)
ax1.bar(df['class_name'], df['f1_score'], color=color, alpha=0.7)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(df['class_name'], rotation=45, ha="right")

# Create a second y-axis to plot ROC AUC as a line
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('ROC AUC (Line)', color=color)
ax2.plot(df['class_name'], df['roc_auc'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()

# Reshape the data for grouped bar plotting
df_melted = df.melt(id_vars='class_name', value_vars=['precision', 'recall'], var_name='Metric', value_name='Score')

# Create the plot
plt.figure(figsize=(14, 8))
ax = sns.barplot(x='class_name', y='Score', hue='Metric', data=df_melted, palette='viridis')

# Set titles and labels
ax.set_title('Precision and Recall by Class', fontsize=16, pad=20)
ax.set_xlabel('Class Name')
ax.set_ylabel('Score')
ax.set_ylim(0, 1.1)
plt.xticks(rotation=45, ha="right")
plt.legend(title='Metric')
plt.tight_layout()
plt.show()

# Select only the confusion matrix columns
df_confusion = df.set_index('class_name')[['true_positives', 'false_negatives', 'true_negatives', 'false_positives']]

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df_confusion, annot=True, cmap="YlGnBu", fmt='g') # 'fmt="g"' prevents scientific notation

# Set titles and labels
plt.title('Error map', fontsize=16, pad=20)
plt.xlabel('Prediction Outcome')
plt.ylabel('Class Name')
plt.show()
```
